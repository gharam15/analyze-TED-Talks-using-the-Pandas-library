```python
# turn to lowercase to uppercase
class alphabet:
  def __init__(self,nstr) :
    self.lenght=len(nstr)
    self.nstr=nstr
  def uppercase (self):
      somestr=self.nstr
      my_list_str=[]
      for letter in somestr:
        #ascii code
       if 97 <=  ord(letter) <=122:
         my_list_str.append(chr(ord(letter)-32))
      else:
         my_list_str.append(letter)

      return ''.join(my_list_str)
```
